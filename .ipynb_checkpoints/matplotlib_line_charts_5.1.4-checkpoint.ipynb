```python
# import dependencies
import matplotlib.pyplot as plt
import statistics as stat
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.plot(months,savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Saving Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
plt.xlabel('Month')
plt.ylabel('Fares Accumlated ($)')
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.show()
```

```python
# Set figure size to 20x10 pixels
plt.figure(figsize=(20,10))

# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()

plt.show()
```

```python
# create plot with: Green line, diamond markers, 2pt linewidth, legend=Boston
plt.plot(x_axis,y_axis, marker='D', color='Green', linewidth=2, label='Boston')
#Add title and axes labels
plt.xlabel('Date')
plt.ylabel('Fares ($)')

#Add y-axis limit of 45
plt.ylim(0,45)

#Add grid lines
plt.grid()

plt.legend()

plt.show()
```

```python
stdev=stat.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis,y_axis,yerr=stdev)
```

```python
fig,ax=plt.subplots()
ax.errorbar(x_axis,y_axis,yerr=stdev,capsize=10)
# capsize makes the lines into I shapes
plt.show()
```
