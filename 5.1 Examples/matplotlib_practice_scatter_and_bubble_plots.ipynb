```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(y_axis, x_axis, 'o', color='red', label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Alternative way of increasing size of points in scatter plot
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis
           # change bubble to 20% transparent skyblue
           ,color='skyblue',alpha=0.8
           # add black border with width of 2
           ,edgecolor='k',linewidths=2
           # adjust bubble size to 5*y-axis (the list y-axis, which is on the x-axis of this chart)
           ,s = [5*i for i in y_axis]
          , label='Boston')

ax.invert_yaxis()
ax.set_xlim(0,50)
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()

plt.show()
```
